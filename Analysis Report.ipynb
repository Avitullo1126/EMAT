To make the graphs easier to read I am going to be using the index_col command to show the danceability first and luckily energy is not far down. 

```python
audio_features = pd.read_csv('one_more_time.csv', index_col=['Danceability'])
audio_features.head()
```

```python
audio_features2 = pd.read_csv('wolf.csv', index_col=['Danceability'])
audio_features2.head()
```

```python
mid_df = audio_features2[audio_features2['Album Name'] == 'wolf']
lon_df = audio_features2[audio_features2['Album Name'] == 'one more time']

fig5, ax5 = plt.subplots()
ax5.scatter(x = audio_features2['Energy'],
            y = audio_features2['Danceability'])
ax5.set_title('wolf')
ax5.set_xlabel('Danceability')
ax5.set_ylabel('Energy')


fig5, ax5 = plt.subplots()
ax5.scatter(x = audio_features['Danceability'],
            y = audio_features['Energy'])
            
ax5.set_title('one more time')
ax5.set_xlabel('Danceability')
ax5.set_ylabel('Energy')
plt.show()
```
